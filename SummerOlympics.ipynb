### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df1 = df[df['Medal'] == 'Gold']
df1
```

```python
df1['Discipline'].unique()
```

```python
len(df1['Discipline'].unique())
```

```python
for Discipline in df1['Discipline'].unique():
    print([Discipline , len(df1[df1['Discipline']  == Discipline])])
```

```python
data = []
for Discipline in df1['Discipline'].unique():
    data.append([Discipline , len(df1[df1['Discipline']  == Discipline])])
    
data = pd.DataFrame(data, columns = ['Discipline','No. of Gold Medal'])
```

```python
data.sort_values(by = 'No. of Gold Medal', ascending = False).head()
```

```python
pd.DataFrame(data,columns = ['Discipline','No. of Gold Medal']).sort_values(by='No. of Gold Medal', ascending=False).head().plot(x = 'Discipline', y = 'No. of Gold Medal', kind = 'bar', figsize = (8,6))
```
